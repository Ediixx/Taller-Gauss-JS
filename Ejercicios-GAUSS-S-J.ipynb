Ejercicio 1

Gauss-Jacobi

```python
%autoreload 2
from src import gauss_jacobi

A = [[2,10],[3,2]]

b = [16, 11]
x0 = [1,1]
tol = 1e-6
max_iter = 100

x,tray_jacobi = gauss_jacobi(A=A, b=b, x0=x0, tol=tol, max_iter=max_iter)

print("Solución:\n", x,tray_jacobi)
```

```python
import matplotlib.pyplot as plt

# Extraer las trayectorias de x[0] y x[1]
x0_vals_J = [xi[0][0] for xi in tray_jacobi]
y0_vals_J = [xi[1][0] for xi in tray_jacobi]

# Graficar las iteraciones, el punto inicial y el punto final
plt.figure(figsize=(8, 6))

# Graficar la trayectoria de iteraciones
plt.plot(x0_vals_J, y0_vals_J, linestyle='--', marker='o', label='Iteraciones')

# Resaltar el punto inicial
plt.scatter(tray_jacobi[0][0][0], tray_jacobi[0][1][0], color='green', marker='^', s=150, label='Punto inicial', edgecolors='black', linewidths=2)

# Resaltar el punto final
plt.scatter(x[0][0], x[1][0], color='purple', marker='s', s=150, label='Solución final', edgecolors='black', linewidths=2)

# Configurar etiquetas y título
plt.xlabel('x[0]')
plt.ylabel('x[1]')
plt.title('Grafica Gauss-Jacobi')
plt.grid(True)
plt.legend()
plt.show()
```
