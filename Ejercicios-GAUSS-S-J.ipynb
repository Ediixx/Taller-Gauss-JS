## CONVERGENCIA

Primero con x0=[1,1] 

Gauss-Jacobi y Seidel

```python
A = np.array([[3, 2], [2, 10]], dtype=float)
b = np.array([11, 16], dtype=float)
x0 = np.array([[1], [1]])
tol = 1e-5
max_iter = 100

# Resolver el sistema con Gauss-Jacobi
x_jacobi, tray_jacobi = gauss_jacobi(A=A, b=b, x0=x0, tol=tol, max_iter=max_iter)
tray_jacobi = np.squeeze(np.array(tray_jacobi))  # Convertir trayectoria a numpy
x1_jacobi = tray_jacobi[:, 0] 
x2_jacobi = tray_jacobi[:, 1] 

# Resolver el sistema con Gauss-Seidel
x_seidel, tray_seidel = gauss_seidel(A=A, b=b, x0=x0, tol=tol, max_iter=max_iter)
tray_seidel = np.squeeze(np.array(tray_seidel))  # Convertir trayectoria a numpy
x3_seidel = tray_seidel[:, 0] 
x4_seidel = tray_seidel[:, 1]  
# Graficar las trayectorias
plt.plot(x1_jacobi, x2_jacobi, marker='o',color = 'green', label="Trayectoria Gauss-Jacobi", linestyle='-')
plt.plot(x3_seidel, x4_seidel, marker='s',color = 'yellow', label="Trayectoria Gauss-Seidel", linestyle='--')

# Personalizar la gráfica
plt.title("Gauss-Jacobi vs Gauss-Seidel")
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.grid()
plt.legend()
plt.show()
```
